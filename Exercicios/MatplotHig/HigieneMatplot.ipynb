```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mÃªs
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
total_por_produto = total_por_mes = creme_facial + limpeza_facial + pasta_dentaria + sabonete + shampoo + hidratante

```

```python
# GrÃ¡fico 1 - Total de produtos Vendidos por mÃªs - Linha
plt.figure(figsize=(10, 4))
plt.plot(mes, total_por_mes, label='Limpeza Facial', marker='o')
plt.xlabel('MÃªs')
plt.ylabel('Total Vendido por MÃªs')
plt.title('Total Vendido por MÃªs (Todos os Produtos)')
plt.grid(True)
```

```python
# GrÃ¡fico 2 - GrÃ¡fico com todos os produtos vendidos por mÃªs - Linha
plt.figure(figsize=(10, 6))
plt.plot(mes, creme_facial, label='Creme Facial', marker='o')
plt.plot(mes, limpeza_facial, label='Limpeza Facial', marker='o')
plt.plot(mes, pasta_dentaria, label='Pasta Dentaria', marker='o')
plt.plot(mes, sabonete, label='Sabonete', marker='o')
plt.plot(mes, shampoo, label='Shampoo', marker='o')
plt.plot(mes, hidratante, label='Hidratante', marker='o')
plt.xlabel('Mes')
plt.ylabel('Quantidade Vendida')
plt.title('Produtos Vendidos por Mes')
plt.legend()
plt.grid(True)
```

```python
# GrÃ¡fico 3 - Comparativo de Creme Facial com Limpeza Facial por mÃªs - Barras
plt.figure(figsize=(10, 4))
plt.bar(mes, creme_facial, label='Creme Facial', alpha=0.7)
plt.bar(mes, limpeza_facial, label='Limpeza Facial', alpha=0.7)
plt.xlabel('MÃªs')
plt.ylabel('Quantidade Vendida')
plt.title('Comparativo de Creme Facial com Limpeza Facial por MÃªs')
plt.legend()
plt.grid(True)
```

```python
# GrÃ¡fico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos (1000-1999, 2000-2999, ...)
plt.figure(figsize=(8, 6))
hist_data = [creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante]
plt.hist(hist_data, bins=[1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000],
         label=['Creme Facial', 'Limpeza Facial', 'Pasta Dentaria', 'Sabonete', 'Shampoo', 'Hidratante'],
         alpha=0.7, stacked=True)
plt.xlabel('Quantidade Vendida')
plt.ylabel('NÃºmero de Meses')
plt.title('Histograma de Quantidade de Produtos Vendidos por Mes')
plt.legend()
plt.grid(True)
```

```python
# GrÃ¡fico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto
total_por_produto = np.sum([creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante], axis=1)
labels = ['Creme Facial', 'Limpeza Facial', 'Pasta Dentaria', 'Sabonete', 'Shampoo', 'Hidratante']
plt.figure(figsize=(6, 6))
plt.pie(total_por_produto, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Porcentagem da Qtd de Produtos Vendidos no Ano por Produto')
```
