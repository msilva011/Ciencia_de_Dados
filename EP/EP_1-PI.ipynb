```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Função para calcular a série de Taylor da raiz quadrada
def sqrt_taylor(x, n):
    result = 0.0
    for i in range(n):
        result += ((-1) ** i) * (math.factorial(2 * i) / (4 ** i * math.factorial(i) ** 2)) * (x - 1) ** i
    return result

# Função para calcular as derivadas da raiz quadrada
def sqrt_derivative(x, n):
    if n == 1:
        return 1 / (2 * math.sqrt(x))
    elif n == 2:
        return -1 / (4 * x ** (3/2))
    else:
        return None

# Aproximações e propriedades da função
x_values = np.linspace(0.1, 10, 100)  
y_values_exact = np.sqrt(x_values) 
y_values_taylor = [sqrt_taylor(xi, 5) for xi in x_values]
y_values_derivative = [sqrt_derivative(xi, 2) for xi in x_values]



# Informações adicionais
print("Derivadas da função sqrt(x): y' = 1 / (2 * sqrt(x)), y'' = -1 / (4 * x^(3/2))")
print("O valor de N na série de Taylor representa o grau do polinômio utilizado para a aproximação.")
print("Neste caso, utilizei N=5 para uma boa precisão sem aumentar muito a complexidade computacional.")

# Plotagem dos resultados
plt.figure(figsize=(12, 5))

# Gráfico da função exata vs. aproximação da série de Taylor
plt.subplot(1, 2, 1)
plt.plot(x_values, y_values_exact, label='sqrt(x)', color='blue')
plt.plot(x_values, y_values_taylor, label='Taylor (n=5)', linestyle='--', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função exata vs. Série de Taylor')
plt.legend()

# Gráfico da derivada de segunda ordem
plt.subplot(1, 2, 2)
plt.plot(x_values, y_values_derivative, label="Derivada (n=2)", color='green')
plt.xlabel('x')
plt.ylabel('y\'\'')
plt.title('Derivada de segunda ordem')
plt.legend()

plt.tight_layout()
plt.show()

# Tabela de valores fixos
table_values = [0.5, 1, 2, 5, 10]
table_results = {'x': table_values, 'sqrt(x)': np.sqrt(table_values)}
print("Tabela de Valores Fixos:")
for i in range(len(table_values)):
    print(f"x: {table_values[i]}, sqrt(x): {table_results['sqrt(x)'][i]}")
```
